```python
#____________________________________________________________________________________________________________________________

#--------------------------------------------------Recomendation Based On Genre, Top Rating-----------------------------------------------------------------------------
#_______________________________________________________________________________________________________________________________________________________#


import requests

def get_genre_ids(api_key, genre_names):
    # TMDb API endpoint for movie genres
    url = f'https://api.themoviedb.org/3/genre/movie/list?api_key={api_key}&language=en-US'

    try:
        # Make the API request
        response = requests.get(url)
        response.raise_for_status()

        # Parse the JSON response
        genre_data = response.json()

        # Find the genre IDs for the given genre names
        genre_ids = []
        for genre_name in genre_names:
            for genre in genre_data['genres']:
                if genre['name'].lower() == genre_name.lower():
                    genre_ids.append(genre['id'])
                    break

        return genre_ids

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An unexpected error occurred: {err}")

def get_top_rated_movies(api_key, genre_ids=None):
    # TMDb API endpoint for top rated movies
    url = f'https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page=1'
    
    if genre_ids:
        # If genre_ids are provided, append them to the API request
        genre_ids_str = ','.join(map(str, genre_ids))
        url += f'&with_genres={genre_ids_str}'

    try:
        # Make the API request
        response = requests.get(url)
        response.raise_for_status()

        # Parse the JSON response
        movie_data = response.json()

        # Display the titles of top rated movies
        print(f"Top 20 Rated Movies:")
        for i, movie in enumerate(movie_data['results'][:20]):
            print(f"{i + 1}. {movie['title']}")

        while True:
            # Ask the user for their choice
            user_choice = input("Enter the index of the movie you want to know more about (or enter 'None' to end): ")

            # Process the user's choice
            if user_choice.lower() == 'none':
                print("Thank you for using the movie recommendation service.")
                break
            elif user_choice.isdigit() and 1 <= int(user_choice) <= min(20, len(movie_data['results'])):
                selected_movie_index = int(user_choice) - 1
                selected_movie_title = movie_data['results'][selected_movie_index]['title']
                get_movie_details(api_key, selected_movie_title)
            else:
                print("Invalid choice. Please enter a valid index or 'None' to end.")

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An unexpected error occurred: {err}")

def get_movie_details(api_key, movie_title):
    # TMDb API endpoint for movie details by title
    url = f'https://api.themoviedb.org/3/search/movie?api_key={api_key}&query={movie_title}&language=en-US&page=1&include_adult=false'

    try:
        # Make the API request
        response = requests.get(url)
        response.raise_for_status()

        # Parse the JSON response
        movie_data = response.json()

        # Display relevant information
        if movie_data['total_results'] > 0:
            selected_movie = movie_data['results'][0]
            print(f"\nDetails for '{selected_movie['title']}':")
            print(f"Release Date: {selected_movie['release_date']}")
            print(f"Overview: {selected_movie['overview']}")
            print(f"Rating: {selected_movie['vote_average']}")  # Added rating information
        else:
            print(f"No details found for '{movie_title}'.")

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An unexpected error occurred: {err}")

# Replace 'YOUR_API_KEY' with your actual TMDb API key
#api_key = 'YOUR_API_KEY'

# Get a list of genres from the user (e.g., "Action, Drama, Comedy")
genre_input = input("Enter a list of genres separated by commas: ")
genres = [genre.strip() for genre in genre_input.split(',')]

# Get the genre IDs for the specified genre names
genre_ids = get_genre_ids(api_key, genres)

# Get top rated movies for the chosen genres
get_top_rated_movies(api_key, genre_ids)
```

```python
#______________________________________________________________________________________________________________________________

#--------------------------------------------------Recomendation Based On Genre, Top Rating, Year Range-----------------------------------------------------------------------------
#_______________________________________________________________________________________________________________________________________________________#

import requests
from datetime import datetime

def get_genre_ids(api_key, genre_names):
    # TMDb API endpoint for movie genres
    url = f'https://api.themoviedb.org/3/genre/movie/list?api_key={api_key}&language=en-US'

    try:
        # Make the API request
        response = requests.get(url)
        response.raise_for_status()

        # Parse the JSON response
        genre_data = response.json()

        # Find the genre IDs for the given genre names
        genre_ids = []
        for genre_name in genre_names:
            for genre in genre_data['genres']:
                if genre['name'].lower() == genre_name.lower():
                    genre_ids.append(genre['id'])
                    break

        return genre_ids

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An unexpected error occurred: {err}")

def get_top_rated_movies(api_key, genre_ids=None, start_year=None, end_year=None):
    # TMDb API endpoint for top rated movies within a specified release year range
    url = f'https://api.themoviedb.org/3/discover/movie?api_key={api_key}&language=en-US&sort_by=vote_average.desc&vote_count.gte=500'

    if genre_ids:
        # If genre_ids are provided, append them to the API request
        genre_ids_str = ','.join(map(str, genre_ids))
        url += f'&with_genres={genre_ids_str}'

    try:
        # Make the API request
        response = requests.get(url)
        response.raise_for_status()

        # Parse the JSON response
        movie_data = response.json()

        # Filter movies based on release year range
        if start_year is not None and end_year is not None:
            movie_data['results'] = [movie for movie in movie_data['results'] if start_year <= int(movie['release_date'][:4]) <= end_year]

        # Display the titles of top rated movies
        print(f"Top Rated Movies in the specified release year range:")
        for i, movie in enumerate(movie_data['results'][:20]):
            print(f"{i + 1}. {movie['title']} ({movie['release_date'][:4]})")

        while True:
            # Ask the user for their choice
            user_choice = input("Enter the index of the movie you want to know more about (or enter 'None' to end): ")

            # Process the user's choice
            if user_choice.lower() == 'none':
                print("Thank you for using the movie recommendation service.")
                break
            elif user_choice.isdigit() and 1 <= int(user_choice) <= min(20, len(movie_data['results'])):
                selected_movie_index = int(user_choice) - 1
                selected_movie_title = movie_data['results'][selected_movie_index]['title']
                get_movie_details(api_key, selected_movie_title)
            else:
                print("Invalid choice. Please enter a valid index or 'None' to end.")

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An unexpected error occurred: {err}")

def get_movie_details(api_key, movie_title):
    # TMDb API endpoint for movie details by title
    url = f'https://api.themoviedb.org/3/search/movie?api_key={api_key}&query={movie_title}&language=en-US&page=1&include_adult=false'

    try:
        # Make the API request
        response = requests.get(url)
        response.raise_for_status()

        # Parse the JSON response
        movie_data = response.json()

        # Display relevant information
        if movie_data['total_results'] > 0:
            selected_movie = movie_data['results'][0]
            print(f"\nDetails for '{selected_movie['title']}':")
            print(f"Release Date: {selected_movie['release_date']}")
            print(f"Rating: {selected_movie['vote_average']}")
            print(f"Overview: {selected_movie['overview']}")
        else:
            print(f"No details found for '{movie_title}'.")

    except requests.exceptions.HTTPError as errh:
        print(f"HTTP Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"An unexpected error occurred: {err}")

# Replace 'YOUR_API_KEY' with your actual TMDb API key
api_key = '4d309e290983c8b9608254a9f3a9b7db'

# Get a list of genres from the user (e.g., "Action, Drama, Comedy")
genre_input = input("Enter a list of genres separated by commas: ")
genres = [genre.strip() for genre in genre_input.split(',')]

# Get the genre IDs for the specified genre names
genre_ids = get_genre_ids(api_key, genres)

# Get the release year range from the user
start_year = input("Enter the start year for the movie range (e.g., 2000): ")
end_year = datetime.now().year  # Use the current year as the end of the range

# Get top rated movies for the chosen genres and release year range
get_top_rated_movies(api_key, genre_ids, int(start_year), end_year)
```
